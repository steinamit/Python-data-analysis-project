## Does academic status indicate safer driving ?

#### This analysis aims to examine how academic status relates to the type of driver's license held and the driving behavior of the individual being examined.

```python
value_counts_Educ = pps['Educ'].value_counts()
value_counts_Educ
```

```python
filtered_pps = pps[~pps['Educ'].isin(['Irrelevant', 'Unknown','Non', '--','Other'])]

explode = [0.2, 0.2, 0.2, 0,0,0]

filtered_pps['Educ'].value_counts().plot.pie(autopct='%.0f%%', figsize=(15, 6), explode = explode, shadow=True, rotatelabels='true');
plt.title('spread of education')
```

####  The top 3 dominant groups of the respondents:
#### 1. First Degree 
#### 2. High School
#### 3. High School no dp


#### This data helps to understand the reliability of each category we tested, based on their dominance in the sample.

#### In any case, the data makes sense because those with more advanced degrees are more rare than for example a high school education or a bachelor's degree. Therefore, these data can be used for the purpose of the research presented here.

```python
pps_filtered = pps[~pps['Educ'].isin(['Irrelevant', 'Unknown','Non', '--','Other'])]

plt.figure(figsize=(15, 6))
sns.boxplot(x='Educ', y='Age', data=pps_filtered)

plt.title('Boxplot of Age by Education')
plt.xlabel('Education')
plt.ylabel('Age')
plt.show()
```

### For the accuracy of the data, we will analyze the abnormal ages

#### 1.First Degree,Second Degree: There are outliers.
#### 2.Others: No outliers detected.

#### the age range in all the columns almost overlaps, with the exception of "High School" and "Thired Degree".

### We will filter out the outliers

```python
education_types_corrected = ['First.Degree', 'No Acadmic Degree', 'High.school.No Dp', 'High.school', 'Second.Degree', 'Thired.Degree']

pps_cleaned = pd.DataFrame()

for ed_type in education_types_corrected:
    subset = pps_filtered[pps_filtered['Educ'].str.strip() == ed_type.strip()]
    
    if not subset.empty:
        Q1 = subset['Age'].quantile(0.25)
        Q3 = subset['Age'].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        cleaned_subset = subset[(subset['Age'] >= lower_bound) & (subset['Age'] <= upper_bound)]
        
        if not cleaned_subset.empty:
            pps_cleaned = pd.concat([pps_cleaned, cleaned_subset], ignore_index=True)
        else:
            print(f"Warning: No data after cleaning for {ed_type}")
    else:
        print(f"Warning: No data for {ed_type} before cleaning")

print(f"Number of rows before cleaning: {len(pps_filtered)}")
print(f"Number of rows after cleaning: {len(pps_cleaned)}")

print(len(pps_filtered)-len(pps_cleaned))
```

```python
plt.figure(figsize=(15, 6))
sns.boxplot(x='Educ', y='Age', data=pps_cleaned)

plt.title('Boxplot of Age by Education After filtering')
plt.xlabel('Education')
plt.ylabel('Age')
plt.show()
```

#### The filtering we did will help us get more reliable data.

```python
filtered_pps = pps_cleaned[~pps_cleaned['Educ'].isin(['Irrelevant', 'Unknown', '--', 'Other']) & 
                   ~pps_cleaned['profDrvr'].isin(['Unknown', '--'])]

pps_pivot = filtered_pps.pivot_table(values='Age', index='Educ', columns='profDrvr', aggfunc='count')
pps_pivot
```

```python
pps_pivot_percentage = round(pps_pivot.div(pps_pivot.sum(axis=1), axis=0) * 100,1)
ax=sns.barplot(data=pps_pivot_percentage.reset_index(), x='Educ', y='No', color='skyblue', label='No')
ax=sns.barplot(data=pps_pivot_percentage.reset_index(), x='Educ', y='Yes', color='orange', label='Yes')
ax.bar_label(ax.containers[1], labels=[f'{val.get_height():.1f}%' for val in ax.containers[1]], fontsize=15)
plt.title('Percentage of profDrvr by Educ')
plt.xlabel('Educ')
plt.ylabel('Percentage')
plt.xticks(rotation=45)
plt.legend(title='profDrvr')
plt.tight_layout()
plt.show()
```

* The data is shown as an approximation

1. High School (with or without Dp): 7% safer driving habits
2. No Academic Degree: 6% safer driving habits
3. First Degree: 3% safer driving habits
4. Second Degree: 2% safer driving habits
5. Third Degree: 1% safer driving habits


People across all Educational levels present low percentages of safer driving habits , the lowest observed among those with a Third Degree.

## `Observation 1`

### Based on the results, there appears to be **negative correlation** between academic status and safe driving, people with lower education status seem to exhibit safer driving  habits according to the findings.

##### In order to analysiss the gender and the sector columns, we need to merge the datasets - **pps** and **hhs**.

```python
pps['HHID2']=pps['HHID2'].astype(str) 

hhs['index']=hhs['HHID']
pps['index']=pps['HHID2']
new_table1=pps.merge(hhs,left_on='index',right_on='index',how='left')
new_table1
```

```python
filtered_analysis = new_table1[~new_table1['CLic'].isin(["Unknown",'Irrelevant','--'])] 
```

```python
analysis=filtered_analysis.groupby(['Gender','Sector','CLic']).agg({'CLic':'count'}).rename(columns={'CLic': 'Count'})

percentages = []


for i in range(0, len(analysis['Count']), 2):
    count1 = analysis['Count'][i]
    count2 = analysis['Count'][i + 1] if i + 1 < len(analysis) else 0  
    

    total_count = count1 + count2
    
 
    if total_count > 0:
        percentage1 = (count1 / total_count) * 100
        percentage2 = (count2 / total_count) * 100
    else:
        percentage1 = 0
        percentage2 = 0
    
 
    percentages.append(percentage1)
    percentages.append(percentage2)


analysis['Percentage'] = percentages

analysis
```

```python
plt.figure(figsize=(12, 8))

ax = analysis['Percentage'].unstack().plot(kind='bar', width=0.8, color=['skyblue', 'salmon'])
plt.title('Percentage of Car License by Gender and Sector', fontsize=18)
plt.xlabel('Gender and Sector', fontsize=15)
plt.ylabel('Percentage', fontsize=15)

plt.grid(axis='y', linestyle='--', alpha=0.5)
ax.bar_label(ax.containers[0], labels=[f'{x:.2f}%' for x in ax.containers[0].datavalues], fontsize=8)
ax.bar_label(ax.containers[1], labels=[f'{x:.2f}%' for x in ax.containers[1].datavalues], fontsize=8)

plt.xticks(rotation=70)
plt.legend(title='Car License', fontsize=12, loc='upper right')
plt.tight_layout()
plt.show()
```
